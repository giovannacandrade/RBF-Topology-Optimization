```python
from dolfin import *
from matplotlib import cm,pyplot as pp
import numpy as np, sys, os
import time
import numpy.matlib
import scipy
%matplotlib inline
%run plsf.py
%run structure.py
%run aux.py

# ---- Structure parameters --------------------------
name = 'cantilever1'
eps_er, E, nu = [1e-3, 1.0, 0.3]; lx, ly = [2.0, 1.0]; NSx, NSy = [180,90]; nNodesS = (NSx+1)*(NSy+1) + NSx*NSy
velocity_ext = 'HB' 
# ---- PLSF parameters ----------------------
NRx, NRy = [60,30]; nNodesR = (NRx+1)*(NRy+1)
dsp = (2.5)*(lx/30) 
# ---- Optimization parameters -----------------------
Lag, rho, rho_max = [60.0,1.0, 5e3]; gamma = 1.2; tau = 0.9; inv=[]
hS = (ly/NSy); hR = (ly/NRy)
dt = np.minimum(0.02, 0.95*hR); eps = 1.0*hS
reinit = False
zeta = lx; volfrac = 0.5
MG = (2.0)*np.sqrt(hR*hS)
a = np.minimum(1.25*10**(-3), np.maximum(MG**2, 7.5*10**(-4)))

# ---- PLSF Initialization ---------------------------
XR,YR = np.meshgrid(np.linspace(0.0,lx,NRx+1),np.linspace(0.0,ly,NRy+1)); 
r = 0.1*ly 
hX = lx*np.array([1/6, 5/6,1/6,5/6,1/6,5/6,0,1/3,2/3,1,0,1/3,2/3,1,1/2])
hY = ly*np.array([0.,0.,0.5,0.5,1.,1.,0.25,0.25,0.25,0.25,0.75,0.75,0.75,0.75,0.5])
Phi = np.sqrt((XR-hX[0])**2+(YR-hY[0])**2)-r
for iH in range(1,len(hX)):
    Phi = np.minimum(Phi,np.sqrt((XR-hX[iH])**2+(YR - hY[iH])**2)-r);
end
phi_mat = np.maximum(-0.1,np.minimum(0.1,Phi)) 

plsf = PLSF(phi_mat, dsp, lx, ly, XR, YR, nNodesR)

# ---- Structure Initialization ----------------------
structure = Structure(name, lx, ly, NSx, NSy, E, nu, eps_er, velocity_ext,a)
XS = structure.dofsV[:,0]; YS = structure.dofsV[:,1]
GT,prpX1,prpX2 = plsf.transf_matrix(XR.flatten("F"),YR.flatten("F"),nNodesR,XS.flatten("F"),YS.flatten("F"),nNodesS,1)
phiT = csr_matrix.dot(GT,plsf.alpha)
heav = Heav(phiT,eps); delta =Delta(phiT,eps); Grad = plsf_grad(plsf, prpX1, prpX2)
structure._map_geometry(heav, delta, Grad)
structure.SolveLinearElasticity()

# # ---- Optimization Initialization -------------------
ItMax,It,stop = [150,1,False] 
Lag_it = [Lag]; rho_it = [rho]
Comp = [structure._get_compliance()]
Vol = [structure._get_volume()]
Jit = [Comp[0] + rho/2*(np.maximum(0,zeta*(Vol[0] - volfrac)+Lag/rho)**2-Lag**2/rho**2)]
dJ = []; step = []; V = np.zeros(ItMax)

# ---- Initial Plots --------------------------------
print('************ Structure Information  ************') 
print('Number of knots   : %s' % nNodesS)
print('NSx, NSy          : %s, %s' % (NSx, NSy))
print('lx, ly            : %.2f, %.2f' % (lx, ly))
print('Velocity extension:', velocity_ext)
if velocity_ext == 'HB'  or velocity_ext == 'HV':
    print('a             : %.4f' % structure.a)
    print('b             : %.2f' % structure.b)
    print('c             : %.2f' % structure.c)
print('************ PLSF Information  ************') 
print('Number of RBFs    : %s' % nNodesR)
print('Support ratio     : %.2f' % plsf.dsp)
print('Inner iterations  : %s' % plsf.inner)
print('************ Optimization Information  ****') 
print('Initial Lag       :', Lag)
print('Initial rho       :', rho)
print('Maximum rho       :', rho_max)
print('************ INITIAL DESIGN  ************')                   
print('Cost Functional       : %.2f' % Jit[0])
print('Compliance            : %.2f' % Comp[0])
print('Volume fraction       : %.2f' % Vol[0]) 

if velocity_ext == 'HB':
    I, CX, CY = _to_Rmesh(plsf, structure,NSx,NSy,NRx,NRy,velocity_ext)
    vel = np.zeros((NRy+1,NRx+1))
# 
##### ---------- MAIN LOOP ---------------------------------------------------
while It < ItMax and stop == False:
    # --- Compute the descent direction th ------------------------------    
    structure.ShapeDerivative(Lag, rho, volfrac, velocity_ext,zeta)
    if velocity_ext == 'HB':
        th_array = structure.Vn.vector().get_local()
        vel[CY,CX] = th_array[I]
        vel =  vel/(np.max(np.max(abs(vel))))
    elif velocity_ext == 'HV':
        vel = _to_Rmesh(plsf,structure,NSx,NSy,NRx,NRy,velocity_ext)
        vel =  vel/(np.max(abs(vel[0]) + abs(vel[1])))      
    elif velocity_ext == 'natural':
        nodeComp = _to_Rmesh(plsf,structure,NSx,NSy,NRx,NRy, velocity_ext)
        vel =  (nodeComp - rho*np.maximum(0,Lag/rho + zeta*(structure.vol - volfrac)))
        vel =  vel/(np.max(np.max(abs(vel)))) #vel/(np.median(abs(vel)))

# --- Update PLSF ---------------------------------------------------
    plsf.evolve(vel, lx, ly, NRx, NRy, dt, velocity_ext) #lsf changes
    phiT = csr_matrix.dot(GT,plsf.alpha) #maps new geometry
    heav = Heav(phiT,eps)
    delta =Delta(phiT,eps)
    Grad = plsf_grad(plsf, prpX1, prpX2)
    structure._map_geometry(heav, delta, Grad)
    #--- Reinit ------------------------------------------------
    if reinit == True and np.mod(It,1) == 0:
        reinit_aprox(plsf, structure)  
        phiT = csr_matrix.dot(GT,plsf.alpha)
        heav = Heav(phiT,eps)
        delta =Delta(phiT,eps)
        Grad = plsf_grad(plsf, prpX1, prpX2)
        structure._map_geometry(heav, delta, Grad)
        mean = assemble(structure.grad*structure.grad*structure.delta*(structure.dX+structure.ds))/assemble(structure.grad*structure.delta*(structure.dX+structure.ds))
        plsf.alpha = plsf.alpha/mean
    #--- Volume Calculus ------------------------------------------------
    vol = structure._get_volume()
    Vol.append(vol)
    # --- Cost Functional ----------------------------------------------- 
    structure.SolveLinearElasticity()
    Comp.append(structure._get_compliance()) 
    Jit.append(Comp[It] + rho/2*(np.maximum(0,zeta*(Vol[It] - volfrac)+Lag/rho)**2 -Lag**2/rho**2))
    #---- Lagrange Multiplier Update ----------------------------------- 
    C = zeta*(Vol[It] - volfrac)
    V[It] = np.maximum(C, - Lag/rho)
    Lag = Lag + rho*np.maximum(zeta*(Vol[It] - volfrac),-Lag/rho)
    invia = abs(V[It])/abs(V[It-1]); inv.append(invia)
    if It > 1 and abs(V[It]) < tau*abs(V[It-1]):
        rho = rho
    else:
        rho = np.minimum(rho*gamma, rho_max)
    Lag_it.append(Lag); rho_it.append(rho)
    print('************ ITERATION NUMBER %s' % It)                   
    print('Cost Functional       : %.2f' % Jit[It])
    print('Compliance            : %.2f' % Comp[It])
    print('Volume fraction       : %.4f' % Vol[It]) 
    #---- Stopping Criteria --------------------------------------------
    if It>10 and all(abs((Jit[It-6:It]-Jit[It])) < (hS/2)*Jit[It]) and all(abs((np.array(Vol[It-3:It])-volfrac)) < hS):
        stop = True
    #---- Plot Geometry ------------------------------------------------  
    if np.mod(It,10)==0 or It==1 or It==ItMax or stop==True:   
        plsf.plot(It)
    It = It+1
plot_results(Jit, Comp, Vol, It, Lag_it, rho_it, dJ, volfrac, zeta)
```
